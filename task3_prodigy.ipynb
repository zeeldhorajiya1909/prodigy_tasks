```python
import random
from collections import defaultdict

def build_markov_chain(text):
    """Builds a simple word-level Markov chain model."""
    words = text.split()
    chain = defaultdict(list)

    for i in range(len(words) - 1):
        current_word = words[i]
        next_word = words[i + 1]
        chain[current_word].append(next_word)

    return chain

def generate_text(chain, length=20, start_word=None):
    """Generates text using the Markov chain."""
    if not chain:
        return ""

    current_word = start_word if start_word in chain else random.choice(list(chain.keys()))
    result = [current_word]

    for _ in range(length - 1):
        next_words = chain.get(current_word)
        if not next_words:
            break
        current_word = random.choice(next_words)
        result.append(current_word)

    return " ".join(result)

# Example usage:
sample_text = "a man,a plan,a canal:panama"
model = build_markov_chain(sample_text)
generated = generate_text(model, length=10)

print(f"Sample Input: {sample_text}")
print(f"Generated Output: {generated}")
```
